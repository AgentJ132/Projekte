```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import joblib
from sklearn.metrics import classification_report

wine_data = pd.read_csv('wine.csv')  
wine_data = wine_data.dropna()

wine_data_x = wine_data.drop(["Unnamed: 0", "quality", "density", "citric acid", "residual sugar", "free sulfur dioxide", 
                              "total sulfur dioxide", "pH", "sulphates", "alcohol"], axis = 1)

X = wine_data_x.drop(["red_wine"], axis = 1)
y = wine_data_x["red_wine"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

model = KNeighborsClassifier(n_neighbors = 15)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy_score(predictions, y_test)
```
